```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
A = [
    [1,0,0,0],
    [1,0,0,0],
    [1,0,0,0],
    [1,0,0,0]
]

A
```

```python
def create_digraph(adj_matrix):
    graph = nx.DiGraph()
    num_vertices = len(adj_matrix)

    # Iterate through each cell in the adjacency matrix
    for i in range(num_vertices):
        for j in range(num_vertices):
            if adj_matrix[i][j] != 0:
                # Add a directed edge from row vertex to column vertex
                graph.add_edge(i, j)

    return graph
```

```python
# Create a digraph from the adjacency matrix
G = create_digraph(A)

# Visualize the digraph
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.eigenvector_centrality(G)
```
